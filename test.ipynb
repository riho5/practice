###  How to use my package

```python

# and install others in this file 
import numpy as np 
import matplotlib.pyplot as plt
import random as rand
```

```python
# make sample data
t = np.linspace(0,10,200)
y = np.sin(t*2*np.pi)

y_noise = [x + rand.random()*2-1 for x in y]
plt.plot(t,y_noise,label = "include noise"),
plt.plot(t,y,label = "original")

plt.legend()
plt.show()
```

```python
#use lowpass filter
from packages import filters as filt #import package
y_lowpass = filt.get_lowpass(y,0.01)

plt.plot(t,y_noise,label = "include noise"),
plt.plot(t,y_lowpass,label = "lowpass")
plt.plot(t,y,label = "original",alpha = 0.5)

plt.legend()
plt.show()
```
